### Ligistic Regression을 이용한 레드/화이트 와인 분류하기

```
wine = pd.read_csv('./wine.csv', sep = ',', index_col = 0)
wine.head()
```

```
# 레드 와인(1)과 화이트 와인(0) 비율
wine['color'].value_counts()
```

```
# 특성(feature) 데이터(x)와 라벨 데이터(y)의 분리
y = wine['color']
X = wine.drop(['color'], axis = 1)

X.head()
```

```
# 훈련용 데이터와 테스트용 데이터로 나누기
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = \
train_test_split(X, y, test_size = 0.2, random_state = 13)
```

```
# 잘 나눠졌는지 확인
import numpy as np

np.unique(y_train, return_counts = True)
```

```
# Logistic Regression으로 훈련하기
from sklearn.linear_model import LogisticRegression

log_reg = LogisticRegression(random_state = 13, solver = 'liblinear', C = 10)
log_reg.fit(X_train, y_train)
```

```
# 훈련용 데이터의 accuracy 확인
from sklearn.metrics import accuracy_score

pred = log_reg.predict(X_train)
accuracy_score(y_train, pred)
```

```
# 테스트용 데이터의 accuracy 확인
pred = log_reg.predict(X_test)
accuracy_score(y_test, pred)
```

```
# confusion matrix 확인 (0: white, 1: red)
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, pred)
```

```
# bloxplot (12개 특성 중 임의로 선택한 특성 3개)
# 특성들 간 크기 등 너무 차이가 나면 에러를 최적화하는 과정이 험난할 수 있음
import seaborn as sns
sns.boxplot(data = X[['fixed acidity', 'chlorides', 'quality']])
```

```
# normalization : 특성 간 격차를 줄이는 대표적인 방법
  # MinMaxXcaler : 최솟값 0, 최댓값 1이 되도록
  # StandardScaler : 평균과 표준편차가 같아지도록
from sklearn.preprocessing import MinMaxScaler, StandardScaler

MMS = MinMaxScaler()
SS = StandardScaler()

MMS.fit(X)
SS.fit(X)

X_mms = MMS.transform(X)
X_ss = SS.transform(X)
```

```
# 두 방법의 차이를 알기 위해 dataframe화
X_mms_pd = pd.DataFrame(X_mms, columns = X.columns)
X_ss_pd = pd.DataFrame(X_ss, columns = X.columns)
```

```
# MinMaxScaler 결과
sns.boxplot(data = X_mms_pd[['fixed acidity', 'chlorides', 'quality']])
```

```
# StandardScaler 결과
sns.boxplot(data = X_ss_pd[['fixed acidity', 'chlorides', 'quality']])
```

```
# MinMaxScaler로 학습한 결과
X_out = X_mms_pd
X_train, X_test, y_train, y_test = \
train_test_split(X_out, y, test_size = 0.2, random_state = 13)

log_reg = LogisticRegression(random_state = 13, solver = 'liblinear', C = 10.)
log_reg.fit(X_train, y_train)

pred = log_reg.predict(X_test)
accuracy_score(y_test, pred)
```

```
# StandardScaler로 학습한 결과
X_out = X_ss_pd
X_train, X_test, y_train, y_test = \
train_test_split(X_out, y, test_size = 0.2, random_state = 13)

log_reg = LogisticRegression(random_state = 13, solver = 'liblinear', C = 10.)
log_reg.fit(X_train, y_train)

pred = log_reg.predict(X_test)
accuracy_score(y_test, pred)
```

```
# 그런데 어떤 특성이 화이트/레드 와인을 결정하는 걸까?
# Tree 계열 머신러닝은 그 답을 줄 수 있음 - 결정 나무
from sklearn.tree import DecisionTreeClassifier

tree_clf = DecisionTreeClassifier(max_depth = 2, random_state = 13)
tree_clf.fit(X_train, y_train)
```

```
# accuracy는 logistic regression보다는 낮음
pred = tree_clf.predict(X_train)
accuracy_score(y_train, pred)
```

```
pred = tree_clf.predict(X_test)
accuracy_score(y_test, pred)
```

```
# 결정나무 계열은 결과를 얻는 과정을 들여다볼 수 있음
from graphviz import Source
from sklearn.tree import export_graphviz

Source(export_graphviz(tree_clf, feature_names = X_train.columns,
                       class_names = ['W', 'R'],
                       rounded = True, filled = True))
```

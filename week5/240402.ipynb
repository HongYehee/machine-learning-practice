```python
import pandas as pd

# read_excel(파일경로): 엑셀 파일 내용 불러오기
pd.read_excel("./data/teenage_mental.xls")
```

```python
# 엑셀 파일에서 필요한 부분만 읽어올 수 있음
pd.read_excel("./data/teenage_mental.xls", header=1, usecols="C:K")
```

```python
# column 이름 직접 입력하기
col_names = ['스트레스', '스트레스남학생', '스트레스여학생', '우울감경험률', '우울남학생', '우울여학생','자살생각율', '자살남학생', '자살여학생']
pd.read_excel("./data/teenage_mental.xls", header=1, usecols="C:K", names=col_names)
```

```python
# 위의 내용을 변수에 저장하기
raw_data = raw_data = pd.read_excel("./data/teenage_mental.xls", header=1, usecols="C:K", names=col_names)
raw_data
```

```python
# "그렇다" 응답의 반대인 "아니다"의 데이터를 추가하고, 응답 column도 생성해서 추가하기
raw_data.loc[1] = 100. - raw_data.loc[0]
raw_data['응답'] = ['그렇다', '아니다']
raw_data
```

```python
# 응답 column을 index로 변경
raw_data.set_index('응답', drop=True, inplace=True)
raw_data
```

### Matplotlib를 이용한 시각화

```python
# plot.pie(): 데이터를 파이차트로 출력 가능
raw_data['스트레스'].plot.pie()
```

```python
# explode 옵션: 파이차트에 간격 추가 가능
raw_data['스트레스'].plot.pie(explode = [0, 0.02])
```

```python
# set_title(): 제목 추가
# set_ylable(): y축 제목 삭제

f,ax=plt.subplots(1,3, figsize=(16,8))

raw_data['스트레스'].plot.pie(explode=[0,0.02], ax=ax[0], autopct='%1.1f%%')
ax[0].set_title('스트레스를 받은적 있다')
ax[0].set_ylabel('')

plt.show()
```

```python
# sublots(): 3개의 파이차트 출력
f,ax=plt.subplots(1,3, figsize=(16,8))

raw_data['스트레스'].plot.pie(explode=[0,0.02], ax=ax[0], autopct='%1.1f%%')
ax[0].set_title('스트레스를 받은적 있다')
ax[0].set_ylabel('')

raw_data['우울감경험률'].plot.pie(explode=[0,0.02], ax=ax[1], autopct='%1.1f%%')
ax[1].set_title('우울증을 경험한적 있다')
ax[1].set_ylabel('')

raw_data['자살생각율'].plot.pie(explode=[0,0.02], ax=ax[2], autopct='%1.1f%%')
ax[2].set_title('자살을 고민한적 있다')
ax[2].set_ylabel('')

plt.show()
```

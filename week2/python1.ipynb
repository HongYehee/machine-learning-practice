### 3. operator

```python
print(7 + 4) # 덧셈
print(7 - 4) # 뺄셈
print(7 * 4) # 곱셈
print(7 ** 2) # 지수
print(7 // 4) # 나눗셈 (정수 나눗셈의 몫)
print(7 / 4) # 나눗셈 (실수 나눗셈)
print(7 % 4) # 나머지
```

```python
# string
tom = 'boy'
print(tom)
type(tom)
```

```python
# string형에서의 따옴표(quote)
print('"He is a smart boy." My teacher said')
print("He's a smart and diligent boy.")
# 또는 '\' 사용
print("He\'s a smart and diligent boy.")
```

```python
# string concatenation
start = "=========="
title = "Python Program"
finish = "=========="
print(start + title + finish)
```

```python
# len() : string의 길이
# 공백도 포함
str1 = "Hello World"
str2 = "Good Morning"
len(str1) + len(str2)
```

```python
# string의 index
# 공백에도 index가 부여됨
# index는 0부터 시작, 끝에서부터 접근하고 싶으면 -1
str3 = "mountain"
print(str3[0], str3[-1])
```

```python
# string의 slice
tom = "Tom is a good student."
first = tom[2]
second = tom[10]
third = tom[11]
fourth = tom[-5]
print(first + second + third + fourth)
```

```python
# string slice 시 index 범위 지정
fruit = "banana"
print(fruit[:3]) # 처음부터 3을 포함하지 않는 범위까지
print(fruit[3:]) # 3부터 끝까지
print(fruit[3:999]) # 3부터 끝까지
```
